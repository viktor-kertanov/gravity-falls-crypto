```python
import numpy as np
import matplotlib.pyplot as plt
import re
from collections import Counter
```

```python
clean_secret_text = re.sub(r'[^а-яА-Я]', '', secret_text).lower()
clean_normal_text = re.sub(r'[^а-яА-Я]', '', normal_text).lower()
```

```python
secret_data = list(clean_secret_text)
clean_data = list(clean_normal_text)
```

```python
element_counts_secret = Counter(secret_data)
sorted_secrets = sorted(element_counts_secret.items(), key=lambda x: x[1], reverse=True)
elements_secret, counts_secret = zip(*sorted_secrets)
total_count_secret = sum(counts_secret)
relative_count_secret = [el/total_count_secret for el in counts_secret]

element_counts_clean = Counter(clean_data)
sorted_secrets = sorted(element_counts_clean.items(), key=lambda x: x[1], reverse=True)
elements_clean, counts_clean = zip(*sorted_secrets)
total_count_clean = sum(counts_clean)
relative_count_clean = [el/total_count_clean for el in counts_clean]

fig, ax = plt.subplots(1, 2, figsize=(12,4))

ax[0].bar(elements_secret, relative_count_secret)
ax[1].bar(elements_clean, relative_count_clean)

# plt.xlabel('Letters')
# plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```
